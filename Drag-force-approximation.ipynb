```python
from utility import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
Drag = np.loadtxt('Data/2D_DRAG.dat')
Ms = np.loadtxt('Data/2DMs.dat')
Phi = np.loadtxt('Data/2DPhi.dat')
X = np.zeros((64,2))
X[:,0] = Ms
X[:,1] = Phi
print(f"Number of training example: {Drag.shape[0]}")
```

```python
VisualizeData2D(X,Drag,"Drag","Ms","Phi")
```

```python
model = RBFNN(nCentres=40)
model.reset_parameter(X,Drag,beta = 3)
pred = model.forward(X)
print(f"Loss: {MSELoss(pred,Drag):.5f}")
VisualizeData2D(X,pred,"Predicted Drag","Ms","Phi")
```
